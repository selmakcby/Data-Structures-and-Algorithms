##  Breadth-First Search (BFS)

```python
def find(graph_dic, start_id, end_id):

    #initializing queue
    queue = [start_id] # queue --> bfs 
    parent = {start_id: None} # This dictionary will help us keep track of the path
    visited = {} 

    for node in graph_dic.keys():
        visited[node] = "not visited"

    #initializing it first
    visited[start_id] = "visited"
    
    while queue:
        current_node = queue.pop(0) #taking the first element from the list pop(0)
        
        # If we reach the target node, return the path
        if current_node == end_id:
            path = []

            while current_node is not None: #butun nodelar bitene kadar while loopun icinde devam ediyor. 
                path.append(current_node)
                        # 5         :    4 
                #print( current_node, parent[current_node])
                    #4              #4
                current_node = parent[current_node] # checks the last one in the path and reverse the path 

                        #4        :   3 #bc current node is 4 now, and its value is 3.
                        #this continuous until it reaches the None, {1:None} in graph dic
                #print(current_node, parent[current_node]) 
           
            path.reverse() #am i allowed to use this 
            return path  # We reverse the path to start from start_id, 1: None then return 


        for neighbor in graph_dic[current_node]:
            if visited[neighbor] == "not visited":
                visited[neighbor] = "visited"
                queue.append(neighbor)
                parent[neighbor] = current_node #assigning the child to its psrent to keep track on them 
                #print(parent)

def find_path(start_id, end_id, data):
    graph_dic = {}
    for id, following in data:
        graph_dic[id] = following

    res  = find(graph_dic, start_id, end_id)
    return res

find_path(1, 5, [(1, [2]), (2, [3]),(3, [4]), (4, [5]), (5,[])])
```

## Dijkstra's Algorithm

```python
from typing import List, Tuple

def dijkstra(graph_dic, start_id: int, end_id: int):
    
    shortest_distance ={}
    parent = {}
    visited = {}

    for node in graph_dic:
        shortest_distance[node] = float('inf') 
        parent[node] =  None 
        visited[node] = "not visited"

    shortest_distance[start_id] = 0
    
    while True:

        current_node = None 
        curr_distance = float('inf')

        for node, distance in shortest_distance.items():
            if visited[node] == "not visited" and distance < curr_distance:
                current_node = node
                curr_distance = distance 
        #                       {2: 0.5, 3: 0.5}
        for adjNode, weight in graph_dic[current_node].items(): #{1: {2: 0.5, 3: 0.5}, 2: {3: 0.2, 1: 0.2}, 3: {2: 0.16666666666666666}}
            if visited[adjNode] == "not visited":
                new_distance = curr_distance + weight

                if new_distance < shortest_distance[adjNode]: #finding the shortest distance // updating the list in a way 
                    shortest_distance[adjNode] = new_distance
                    parent[adjNode] = current_node  
        
        if current_node == end_id:  # If we reached the end, reconstruct the path
            path = []
            while current_node:
                path.insert(0, current_node) #putting current node into the path list as FIRST elm. 
                current_node = parent[current_node] #curent node is the new value now, updating the current node 
            return path
        
        visited[current_node] = "visited"

def shortest_path(start_id: int, end_id: int,
                  data: List[Tuple[int, int, List[int]]]
                  ) -> List[int]:
    graph_dic = {}
    
    for id, num_likes, following in data:
        node_weight = 1 / num_likes
        #print(num_likes, node_weight)
        graph_dic[id] = {}
        for follows in following:
            #print(i)
            graph_dic[id][follows] = node_weight
            #{1: {2: 0.5, 3: 0.5}, 2: {3: 0.2, 1: 0.2}, 3: {2: 0.16666666666666666}}
    
    return dijkstra(graph_dic, start_id, end_id)


result = shortest_path(1, 3, [(1, 2, [2, 3]), (2, 5, [3, 1]), (3, 6, [2])])
print(result)
```
